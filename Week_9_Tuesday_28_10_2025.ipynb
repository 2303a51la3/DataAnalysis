```python
# Q1: Simulating Hourly Call Volume

import numpy as np
import matplotlib.pyplot as plt

# Given parameters
lam = 15     # average number of calls per hour
hours = 8    # working day (8 hours)

# Simulate data using Poisson distribution
calls = np.random.poisson(lam=lam, size=hours)

# Calculate mean and variance
mean_calls = np.mean(calls)
var_calls = np.var(calls)

print("Hourly Calls Data:", calls)
print(f"Mean calls per hour: {mean_calls:.2f}")
print(f"Variance: {var_calls:.2f}")

# Plot histogram
plt.figure(figsize=(8,5))
plt.hist(calls, bins=6, color='skyblue', edgecolor='black')
plt.title("Histogram of Hourly Customer Calls (λ=15)")
plt.xlabel("Number of Calls per Hour")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()
```

```python
# Q2: Effect of Mean on Poisson Distribution Shape

import numpy as np
import matplotlib.pyplot as plt

# Different lambda values
lambda_values = [5, 10, 15, 25, 40]
size = 1000  # number of samples

plt.figure(figsize=(14,10))

# Generate subplots for each λ
for i, lam in enumerate(lambda_values, 1):
    data = np.random.poisson(lam, size=size)
    plt.subplot(3, 2, i)
    plt.hist(data, bins=20, color='lightgreen', edgecolor='black', alpha=0.7)
    plt.title(f"Poisson Distribution (λ = {lam})")
    plt.xlabel("Number of Calls")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
