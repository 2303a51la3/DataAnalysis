```python

# STUDENT RESULTS ANALYSIS â€“ Pass/Fail Classification

import numpy as np
import pandas as pd

# 1. Data Creation and Categorization

# Sample dataset: 10 student grades
grades = np.array([67, 45, 88, 50, 39, 72, 54, 33, 90, 47])

# Define a flexible threshold (changeable without rewriting logic)
pass_threshold = 50

# Create categorical labels based on threshold
labels = np.where(grades >= pass_threshold, "Pass", "Fail")

# Convert to pandas categorical (factor-like) data
category = pd.Categorical(labels, categories=["Fail", "Pass"])

# 2. Data Structure Analysis


print("---- DATA STRUCTURE ANALYSIS ----")
print(f"Grades: {grades}")
print(f"Pass Threshold: {pass_threshold}")
print(f"Length of Grades Vector: {len(grades)}")
print(f"Data Type of Grades Vector: {grades.dtype}")
print(f"Data Type of Category Vector: {type(category)}")
print(f"Unique Categories: {category.categories.tolist()}\n")

# 3. Statistical Summary
mean_grade = np.mean(grades)
median_grade = np.median(grades)

# Count passes and fails
pass_count = np.sum(grades >= pass_threshold)
fail_count = np.sum(grades < pass_threshold)

# Calculate pass rate (in %)
pass_rate = (pass_count / len(grades)) * 100

print("---- STATISTICAL SUMMARY ----")
print(f"Mean Grade: {mean_grade:.2f}")
print(f"Median Grade: {median_grade:.2f}")
print(f"Number of Passes: {pass_count}")
print(f"Number of Fails: {fail_count}")
print(f"Pass Rate: {pass_rate:.1f}%\n")

# 4. Interpretation & Adaptive Threshold Test


def classify_students(grades, threshold=50):
    """Automatically classify students for any threshold."""
    labels = np.where(grades >= threshold, "Pass", "Fail")
    category = pd.Categorical(labels, categories=["Fail", "Pass"])
    pass_rate = (np.sum(grades >= threshold) / len(grades)) * 100
    return category, pass_rate

# Example: change threshold to 60 without rewriting logic
new_threshold = 60
new_category, new_pass_rate = classify_students(grades, threshold=new_threshold)

print("---- ADAPTIVE THRESHOLD ANALYSIS ----")
print(f"New Threshold: {new_threshold}")
print(f"Categories (Pass/Fail): {list(new_category)}")
print(f"Pass Rate at {new_threshold}: {new_pass_rate:.1f}%")



```

```python

# Optional: Visualization
import matplotlib.pyplot as plt

# Plot bar chart for Pass vs Fail count
counts = pd.Series(category).value_counts()

plt.figure(figsize=(6,4))
counts.plot(kind='bar', color=['salmon', 'mediumseagreen'])
plt.title(f"Student Pass/Fail Distribution (Threshold = {pass_threshold})")
plt.xlabel("Category")
plt.ylabel("Number of Students")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
