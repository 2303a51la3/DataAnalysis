```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

df = pd.read_csv("/content/age_gender.csv")
success_condition = df['gender'] == 0
p_hat = success_condition.mean()

n = 100
size = 10000
simulated = np.random.binomial(n, p_hat, size=size)

x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p_hat)

mean = n * p_hat
var = n * p_hat * (1 - p_hat)

plt.figure(figsize=(10,6))
plt.hist(simulated, bins=30, density=True, alpha=0.6,
         color="skyblue", edgecolor="black", label="Simulation (hist)")
plt.plot(x, pmf, 'r-', lw=2, label="Theoretical PMF")
plt.axvline(mean, color='k', linestyle='--', label=f"Mean = {mean:.1f}")
plt.title(f"Binomial Simulation from Dataset (n={n}, p={p_hat:.2f})")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.legend()
plt.show()

print(f"Mean = {mean:.2f}, Variance = {var:.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

df = pd.read_csv("/content/age_gender.csv")
success_condition = df['gender'] == 0
p_hat = success_condition.mean()

n = 100
size = 10000

simulated = np.random.binomial(n, p_hat, size=size)
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p_hat)

mean = n * p_hat
var = n * p_hat * (1 - p_hat)

plt.figure(figsize=(10,6))
plt.hist(simulated, bins=30, density=True, alpha=0.6,
         color="skyblue", edgecolor="black", label="Simulation (hist)")
plt.plot(x, pmf, 'r-', lw=2, label="Theoretical PMF")
plt.axvline(mean, color='k', linestyle='--', label=f"Mean = {mean:.1f}")
plt.title(f"On-time Deliveries (n={n}, p={p_hat:.2f}) from Dataset")
plt.xlabel("Number of On-time Deliveries")
plt.ylabel("Probability")
plt.legend()
plt.show()

print(f"Mean = {mean:.2f}, Variance = {var:.2f}")

p_values = [0.3, 0.6, 0.9]

for p in p_values:
    simulated = np.random.binomial(n, p, size=size)
    x = np.arange(0, n+1)
    pmf = binom.pmf(x, n, p)
    mean = n * p
    var = n * p * (1 - p)

    plt.figure(figsize=(10,6))
    plt.hist(simulated, bins=30, density=True, alpha=0.6,
             color="lightgreen", edgecolor="black", label="Simulation (hist)")
    plt.plot(x, pmf, 'r-', lw=2, label="Theoretical PMF")
    plt.axvline(mean, color='k', linestyle='--', label=f"Mean = {mean:.1f}")
    plt.title(f"On-time Deliveries (n={n}, p={p})")
    plt.xlabel("Number of On-time Deliveries")
    plt.ylabel("Probability")
    plt.legend()
    plt.show()

    print(f"p={p}: Mean={mean:.2f}, Variance={var:.2f}")
```
